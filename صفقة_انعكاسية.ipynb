```python
def open_reverse_trade_if_loss(symbol):
    """فتح صفقة انعكاسية إذا كانت الصفقة خاسرة ووصل السعر إلى 24% من المسافة إلى الستوب لوز، مع تحديث الستوب لوز والتك بروفيت."""
    positions = mt5.positions_get(symbol=symbol)

    # تحقق من عدد الصفقات المفتوحة
    if len(positions) >= 2:
        print("عدد الصفقات المفتوحة بلغ الحد الأقصى (2). لن يتم فتح صفقة جديدة.")
        return

    if positions:
        for position in positions:
            if position.profit < 0:  # تحقق مما إذا كانت الصفقة خاسرة
                entry_price = position.price_open
                stop_loss = position.sl  # سعر وقف الخسارة
                take_profit = position.tp  # سعر جني الأرباح
                current_price = mt5.symbol_info_tick(symbol).bid if position.type == 0 else mt5.symbol_info_tick(symbol).ask

                # حساب المسافة بين سعر الدخول وتيكن بروفيت
                distance_tp = take_profit - entry_price if position.type == 0 else entry_price - take_profit

                # حساب الستوب لوز الجديد (36% من المسافة إلى تيكن بروفيت)
                new_stop_loss = take_profit - (0.36 * distance_tp) if position.type == 0 else take_profit + (0.36 * distance_tp)

                # حساب السعر الذي يمثل 24% من المسافة نحو وقف الخسارة
                distance_to_sl = entry_price - stop_loss if position.type == 0 else stop_loss - entry_price
                retrace_price = entry_price - (0.24 * distance_to_sl) if position.type == 0 else entry_price + (0.24 * distance_to_sl)

                # تحقق مما إذا كان السعر الحالي قد وصل إلى 24% من المسافة نحو وقف الخسارة
                if (position.type == 0 and current_price <= retrace_price) or \
                   (position.type == 1 and current_price >= retrace_price):

                    # إعداد الطلب لفتح صفقة انعكاسية
                    hedge_request = {
                        "action": mt5.TRADE_ACTION_DEAL,
                        "symbol": symbol,
                        "volume": position.volume,  # نفس حجم المركز الحالي
                        "type": 1 if position.type == 0 else 0,  # 1 = SELL, 0 = BUY
                        "price": mt5.symbol_info_tick(symbol).ask if position.type == 0 else mt5.symbol_info_tick(symbol).bid,
                        "sl": new_stop_loss,  # ستوب لوز جديد
                        "tp": stop_loss,  # تيك بروفت هو نفس ستوب لوز الصفقة القديمة
                        "magic": 123456,
                        "comment": "Reverse trade due to loss",
                    }

                    # إرسال الطلب
                    result = mt5.order_send(hedge_request)
                    if result.retcode != mt5.TRADE_RETCODE_DONE:
                        print(f"فشل في فتح الصفقة الانعكاسية: {result.retcode}")
                    else:
                        print(f"تم فتح صفقة انعكاسية بنجاح: {hedge_request}")

# استخدام الدالة
open_reverse_trade_if_loss("XAUUSD")
```
